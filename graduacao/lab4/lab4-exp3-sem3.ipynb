```python
##### Dependências #####
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st
import scipy.odr as odr
import pandas as pd
sns.set_style("dark")
```

```python
medicao1 = pd.read_csv("medicao1.txt", sep="\t", header=1, decimal=',', 
                       names=["angulo", "tempo", "intensidade"])

medicao2 = pd.read_csv("medicao2.txt", sep="\t", header=1, decimal=',',
                       names=["polarizador1", "nada",
                              "polarizador2", "tempo",
                              "intensidade", "tempo2", "intensidade2"])
medicao2 = medicao2.drop("nada", 1)
nan_inds = ~medicao2.intensidade2.isnull()
medicao2.intensidade[nan_inds] = medicao2.intensidade2[nan_inds]
medicao2.tempo[nan_inds] = medicao2.tempo2[nan_inds]
medicao2 = medicao2.drop("intensidade2", 1)
medicao2 = medicao2.drop("tempo2", 1)
medicao2 = medicao2[~medicao2.intensidade.isnull()]
```

```python
ax = plt.subplot(111, projection="polar")
ax.plot(medicao1.angulo * np.pi / 180, medicao1.intensidade)
plt.title("Intensidade x ângulo no polarizador 2")
plt.show()
```

```python
plt.plot(medicao2.polarizador1, medicao2.polarizador2, '.')
plt.title("Intensidade x ângulo no polarizador 2")
plt.show()
```
