```python
import numpy as np
import scipy as sp
import scipy.misc as misc
import scipy.fftpack as fft
import matplotlib.pyplot as plt
```

```python
img = misc.imread("ruido.jpg")
```

```python
plt.imshow(img)
plt.show()
```

```python
a = np.fft.fft2(img[:, :, 0])
a = np.fft.fftshift(a)
mag_spec = np.log(np.abs(a))
```

```python
plt.matshow(mag_spec)
plt.colorbar()
plt.show()
```
