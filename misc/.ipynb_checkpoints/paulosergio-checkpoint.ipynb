```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
dat = pd.read_excel("paulosergio.xls", sheetname="corrigido")
```

```python
reg = st.linregress(dat.P1, dat.P2)
x = np.arange(100)
y = x * reg.slope + reg.intercept
plt.scatter(dat.P1, dat.P2)
plt.title("Notas da P1 vs P2, Corr: %.2f" % reg.rvalue)
plt.plot(x, y, 'r')
plt.xlim((0, 100))
plt.ylim((0, 100))
plt.show()
```
