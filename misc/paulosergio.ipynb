```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
dat = pd.read_excel("mat220-provas.xls")
```

```python
inds = (~np.isnan(dat.P1)) & (~np.isnan(dat.P2))
p1 = dat.P1[inds]
p2 = dat.P2[inds]
```

```python
reg = st.linregress(p1, p2)
x = np.arange(100)
y = x * reg.slope + reg.intercept
plt.scatter(p1, p2)
plt.title("Notas da P1 vs P2, Corr: %.2f" % reg.rvalue)
plt.plot(x, y, 'r')
plt.xlim((0, 10))
plt.ylim((0, 10))
plt.show()
```
