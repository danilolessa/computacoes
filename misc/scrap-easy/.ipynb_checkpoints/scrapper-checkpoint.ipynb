```python
import crawler
```

```python
def write_emails(emails, path):
    lst = list(emails)
    out = ""
    for l in lst:
        out += l + ", "
    with open(path, "w") as fid:
        fid.write(out)

def automatize(filter_func, dept, url):
    path = "%s-%s.txt" % (dept, url.split("/")[-1:][0])
    url = [url]
    emails = crawler.mail_scrap(filter_func, url)
    write_emails(list(emails), path)
```

```python
def fap():    
    dept = "fap"
    
    def filter_func(link):
        args = link.split("/")
        if "fap" in args:
            if "membro" in args or "membro-do-departamento" in args:
                return True
        return False

    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/docentes")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/pesquisadores")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/estudantes")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/tecnicos")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/administrativos")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/chefias")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/conselhos")
    automatize(filter_func, dept, "http://portal.if.usp.br/fap/pt-br/membros/comissoes")
```
