```python
#####Dependências iniciais.#####
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as st
sns.set_style("darkgrid")
```

```python
#####Parâmetros iniciais.#####
L = 2     #Lado da imagem 'quadrada' [min]#
D = 2     #Densidade de galáxias de fundo [galáxias/min^2]#
sigma = 2  #Incerteza da Normal#
```

```python
#####Processamento inicial. #####

#Número de linhas
N = int(L * L * D)

#Criar um tabela com N colunas e 6 linhas
tabela = np.zeros((6, N))

tabela[0] = np.arange(N) + 1
tabela[1] = np.random.uniform(low=-10.0, high=10.0, size=N)
tabela[2] = np.random.uniform(low=-10.0, high=10.0, size=N)
tabela[3] = np.random.normal(loc=0.0, scale=sigma, size=N)
tabela[4] = np.random.normal(loc=0.0, scale=sigma, size=N)
tabela[5] = np.sqrt(tabela[1] ** 2 + tabela[2] ** 2)
```

```python
#####Output.#####
table = pd.DataFrame(tabela.T, columns=["Contador", "Uniforme1", "Uniforme2", "Gaussiana1", "Gaussiana2", "conta"])
table
```

```python
plt.plot(table.Contador, table.Uniforme1, 'ro', label="$5\\frac{\\alpha_0}{\\gamma}$")
plt.plot(table.Contador, table.Uniforme2, 'g*', label="$3\\frac{\\mu_0}{\\gamma}$")
plt.title("Distribuição uniforme ($D \\in [-10, 10]$) vs outra uniforme")
plt.suptitle("alguma coisa")
plt.xlabel("Contador")
plt.ylabel("Resultado da distribuição")
plt.legend()
plt.show()
```

```python
import scipy.stats as st

N = int(1e4)
x = np.random.randn(N)
x = np.sort(x)
y = st.norm.pdf(x)
plt.hist(x, bins=20, label="Numeros aleatorios gerados", normed=True)
plt.plot(x, y, 'r-', label="PDF gaussiana")
plt.legend(bbox_to_anchor=(1.15, 1))
plt.title("Distribuição gaussiana, N=%s" % N)
plt.ylabel("Densidade de probabilidade")
plt.xlabel("x")
plt.show()
```
