```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw = pd.read_csv("download.csv", names=["time", "pressure", "humidity", "T1", "T2"]) 
```

```python
tim = (raw.time - raw.time[0]) / (60 * 60)
T1 = raw.T1 - 273.15
T2 = raw.T2 - 273.15
max_t = np.max(tim)
nticks = np.ceil(max_t / 24)
ticks = np.linspace(0, nticks * 24, 2*nticks + 1)
```

```python
plt.plot(tim, T1, 'b,')
plt.plot(tim, T2, 'r,')
plt.xlabel("Horas")
plt.xticks(ticks)
plt.ylabel("Temperatura (ºC)")
plt.show()
```

```python
plt.plot(tim, raw.humidity, 'b,')
plt.xlabel("Horas")
plt.xticks(ticks)
plt.ylabel("Umidade (%)")
plt.show()
```

```python
plt.plot(tim, raw.pressure, ',g')
plt.xlabel("Horas")
plt.xticks(ticks)
plt.ylabel("Pressão (hPa)")
plt.show()
```
