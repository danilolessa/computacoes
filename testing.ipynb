```python
def f(p, pk, n):
    if (type(n) is not int) or n < 1:
        return p
    tp = p + pk - p * pk
    if n == 1:
        return tp
    return f(tp, pk, n - 1)
f = np.vectorize(f)
```

```python
p = 0.01
n = np.arange(0, 19)
y = f(p, p, n)
x = n + 1
plt.plot(x, y, '-*')
plt.xlabel("Anos")
plt.ylabel("Probabilidade")
plt.ylim((0, 1))
plt.show()
```

```python
x, g, A = symbols("x g A", real=True)
solve([g <= 1, g >= 0], g)
```
