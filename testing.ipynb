```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = int(1e4)
x = np.linspace(0, 1, N)
y = x ** 2 + 3 * x + np.random.randn()
```

```python
plt.plot(x, y, '.')
plt.show()
```

```python
def f(p, pk, n):
    if (type(n) is not int) or n < 1:
        return p
    tp = p + pk - p * pk
    if n == 1:
        return tp
    return f(tp, pk, n - 1)
f = np.vectorize(f)
```

```python
p = 0.01
n = np.arange(0, 19)
y = f(p, p, n)
x = n + 1
plt.plot(x, y, '-*')
plt.xlabel("Anos")
plt.ylabel("Probabilidade")
plt.ylim((0, 1))
plt.show()
```
