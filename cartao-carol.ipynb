```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados = pd.read_csv("nubank-2016-11.csv")
```

```python
categorias = np.unique(dados.categoria)
```

```python
cat = []
total = []

for categoria in categorias:
    cat.append(categoria)
    total.append(np.sum(dados[dados.categoria == categoria].amount))

total = np.array(total)
```

```python
plt.pie(total, labels=cat, autopct='%1.1f%%')
plt.show()
```
