```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("modis.csv")
```

```python
dat = raw[raw.ScanTime > 0]
#dat = dat[dat.CloudFraction < 100]
t = (dat.ScanTime - np.min(dat.ScanTime)) / (60 * 60 * 24 * 365)
```

```python
tmin = np.min(t)
tmax = np.max(t)
ttot = tmax - tmin
dt = 1 / 12
n = int(ttot / dt)
tt = []
med = []
mean = []
qtd = []
for i in range(0, n):
    tt.append(i)
    datt = dat[((i * dt) < t) & (t < (i + 1) * dt)]
    qtd.append(len(datt))
    med.append(np.median(datt.CloudFraction))
    mean.append(np.median(datt.CloudFraction))
```

```python
plt.bar(tt, mean)
plt.title("Cloud Fraction em Manacapuru (MYD06_L2)")
plt.xticks(np.arange(26))
plt.ylabel("Cloud Fraction mÃ©dio")
plt.xlabel("Tempo (meses) apÃ³s inicio do projeto")
plt.show()
```
